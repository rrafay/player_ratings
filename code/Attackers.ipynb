```
import pandas as pd
import numpy as np
```

```
attackers = pd.read_excel("Forwards.xlsx")
attackers.head()
```

```

val = attackers["goals_scored"] - attackers["xG"]
attackers["Expected_Goal_Difference"] = val
#xgdiff = attackers.iloc[:, [0,4,9,13]]
val2 = attackers["assists"] - attackers["xA"]
attackers["Expected_Assists_Difference"] = val2
#xadiff = attackers.iloc[:, [0,2,10,14]]
```

```
xG = attackers["Expected_Goal_Difference"]
xA = attackers["Expected_Assists_Difference"]
```

```
#Player ratings
attackers_rating = (xG * 0.5 + xA*0.5) * 100
```

```
attackers["ratings"] = attackers_rating
attackers.head()
```

```
attack_player_ratings = attackers.iloc[:, [0,1,11,14]]
#attack_player_ratings = attackers.iloc[:, [0,1,14]]
attack_player_ratings.sort_values(by="ratings", ascending=False)
```

```
d_data = attack_player_ratings.iloc[:, [3]]
```

```
scaled = (d_data - np.min(d_data)) / (np.max(d_data) - np.min(d_data)) *10
```

```
attack_player_ratings["scaled_rating"] = scaled
```

```
attack_player_ratings.sort_values(by="ratings", ascending=False)
```

```
who = attack_player_ratings.iloc[:, [2]]
who_scored_scaled = (who - np.min(who)) / (np.max(who) - np.min(who))*10
```

```
attack_player_ratings["who_scored_scaled_rating"] = who_scored_scaled
```

```
attack_player_ratings.sort_values(by="ratings", ascending=False)
```

```
diff = attack_player_ratings['scaled_rating'] - attack_player_ratings['who_scored_scaled_rating']
attack_player_ratings["difference_of_ratings"] = diff
attack_player_ratings.iloc[:, [0,3,4,5,6]]
```
